```python
import os

# train/images 폴더 내의 파일 이름 변경
image_folder_path = '/content/drive/MyDrive/0_team_torch2tf/baby_datasets/train/images/'
image_files = sorted(os.listdir(image_folder_path))

for i, file_name in enumerate(image_files, 1):
    src = os.path.join(image_folder_path, file_name)
    dst = os.path.join(image_folder_path, f'{i}.jpg')
    os.rename(src, dst)
    print(f'Renamed {src} to {dst}')

# train/labels 폴더 내의 파일 이름 변경
label_folder_path = '/content/drive/MyDrive/0_team_torch2tf/baby_datasets/train/labels/'
label_files = sorted(os.listdir(label_folder_path))

for i, file_name in enumerate(label_files, 1):
    src = os.path.join(label_folder_path, file_name)
    dst = os.path.join(label_folder_path, f'{i}.txt')
    os.rename(src, dst)
    print(f'Renamed {src} to {dst}')
```

```python
import os

# test/images 폴더 내의 파일 이름 변경
image_folder_path = '/content/drive/MyDrive/0_team_torch2tf/baby_datasets/test/images/'
image_files = sorted(os.listdir(image_folder_path))

for i, file_name in enumerate(image_files, 1):
    src = os.path.join(image_folder_path, file_name)
    dst = os.path.join(image_folder_path, f'{i}.jpg')
    os.rename(src, dst)
    print(f'Renamed {src} to {dst}')

# test/labels 폴더 내의 파일 이름 변경
label_folder_path = '/content/drive/MyDrive/0_team_torch2tf/baby_datasets/test/labels/'
label_files = sorted(os.listdir(label_folder_path))

for i, file_name in enumerate(label_files, 1):
    src = os.path.join(label_folder_path, file_name)
    dst = os.path.join(label_folder_path, f'{i}.txt')
    os.rename(src, dst)
    print(f'Renamed {src} to {dst}')
```
