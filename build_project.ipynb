```python
import tensorflow as tf
from tensorflow.keras import layers, models

def unet_model(input_size=(256, 256, 3)):
    inputs = layers.Input(input_size)

    # Encoder
    c1 = layers.Conv2D(64, 3, activation='relu', padding='same')(inputs)
    c1 = layers.Conv2D(64, 3, activation='relu', padding='same')(c1)
    p1 = layers.MaxPooling2D((2, 2))(c1)

    c2 = layers.Conv2D(128, 3, activation='relu', padding='same')(p1)
    c2 = layers.Conv2D(128, 3, activation='relu', padding='same')(c2)
    p2 = layers.MaxPooling2D((2, 2))(c2)

    c3 = layers.Conv2D(256, 3, activation='relu', padding='same')(p2)
    c3 = layers.Conv2D(256, 3, activation='relu', padding='same')(c3)
    p3 = layers.MaxPooling2D((2, 2))(c3)

    c4 = layers.Conv2D(512, 3, activation='relu', padding='same')(p3)
    c4 = layers.Conv2D(512, 3, activation='relu', padding='same')(c4)
    p4 = layers.MaxPooling2D((2, 2))(c4)

    # Bottleneck
    c5 = layers.Conv2D(1024, 3, activation='relu', padding='same')(p4)
    c5 = layers.Conv2D(1024, 3, activation='relu', padding='same')(c5)

    # Decoder
    u6 = layers.Conv2DTranspose(512, 2, strides=(2, 2), padding='same')(c5)
    u6 = layers.concatenate([u6, c4])
    c6 = layers.Conv2D(512, 3, activation='relu', padding='same')(u6)
    c6 = layers.Conv2D(512, 3, activation='relu', padding='same')(c6)

    u7 = layers.Conv2DTranspose(256, 2, strides=(2, 2), padding='same')(c6)
    u7 = layers.concatenate([u7, c3])
    c7 = layers.Conv2D(256, 3, activation='relu', padding='same')(u7)
    c7 = layers.Conv2D(256, 3, activation='relu', padding='same')(c7)

    u8 = layers.Conv2DTranspose(128, 2, strides=(2, 2), padding='same')(c7)
    u8 = layers.concatenate([u8, c2])
    c8 = layers.Conv2D(128, 3, activation='relu', padding='same')(u8)
    c8 = layers.Conv2D(128, 3, activation='relu', padding='same')(c8)

    u9 = layers.Conv2DTranspose(64, 2, strides=(2, 2), padding='same')(c8)
    u9 = layers.concatenate([u9, c1])
    c9 = layers.Conv2D(64, 3, activation='relu', padding='same')(u9)
    c9 = layers.Conv2D(64, 3, activation='relu', padding='same')(c9)

    outputs = layers.Conv2D(1, 1, activation='sigmoid')(c9)

    model = models.Model(inputs=[inputs], outputs=[outputs])

    return model
```

```python
import os

# 폴더 경로 설정
image_folder = '/content/drive/MyDrive/crack_project/crack_dataset/test/images'
mask_folder = '/content/drive/MyDrive/crack_project/crack_dataset/test/masks'

# .jpg 파일 목록 생성
image_files = set(f.lower() for f in os.listdir(image_folder) if f.lower().endswith('.jpg'))
mask_files = set(f.lower() for f in os.listdir(mask_folder) if f.lower().endswith('.jpg'))

# 마스크 폴더와 이름이 같은 이미지 파일 찾기
common_files = list(image_files.intersection(mask_files))

# 남길 파일 개수 설정 (최대 500개)
files_to_keep_count = min(500, len(common_files))
files_to_keep = set(common_files[:files_to_keep_count])

# 남은 파일 목록 작성
all_image_files = [f.lower() for f in os.listdir(image_folder) if f.lower().endswith('.jpg')]
files_to_delete = [f for f in all_image_files if f not in files_to_keep]

# 파일 삭제
for file_name in files_to_delete:
    file_path = os.path.join(image_folder, file_name)

    # 이미지 파일 삭제
    if os.path.exists(file_path):
        os.remove(file_path)
        print(f"Deleted image file: {file_path}")

# 최종 남아있는 파일 개수 확인
remaining_files = len([f.lower() for f in os.listdir(image_folder) if f.lower().endswith('.jpg')])
print(f"Final number of remaining .jpg files: {remaining_files}")

# 남은 파일 개수가 500개가 되도록 강제로 조정
if remaining_files > 500:
    extra_files = [f.lower() for f in os.listdir(image_folder) if f.lower().endswith('.jpg')][500:]

    for file_name in extra_files:
        file_path = os.path.join(image_folder, file_name)
        if os.path.exists(file_path):
            os.remove(file_path)
            print(f"Force deleted image file: {file_path}")

    # 최종 남아있는 파일 개수 확인
    remaining_files = len([f.lower() for f in os.listdir(image_folder) if f.lower().endswith('.jpg')])
    print(f"Final number of remaining .jpg files after forced adjustment: {remaining_files}")
```

```python
import numpy as np
import cv2
import glob
from tensorflow.keras.preprocessing.image import img_to_array  # 필요한 모듈 추가

def load_image(image_path, target_size=(256, 256)):
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f"Image at {image_path} could not be loaded.")
    img = cv2.resize(img, target_size)
    img = img_to_array(img)
    return img

def load_mask(mask_path, target_size=(256, 256)):
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    if mask is None:
        raise ValueError(f"Mask at {mask_path} could not be loaded.")
    mask = cv2.resize(mask, target_size)
    mask = img_to_array(mask)
    return mask

def load_data(image_folder, mask_folder, target_size=(256, 256)):
    image_paths = glob.glob(f"{image_folder}/*.jpg")
    mask_paths = glob.glob(f"{mask_folder}/*.jpg")

    if not image_paths:
        raise ValueError(f"No images found in {image_folder}")
    if not mask_paths:
        raise ValueError(f"No masks found in {mask_folder}")

    if len(image_paths) != len(mask_paths):
        raise ValueError("Number of images does not match number of masks")

    images = [load_image(path, target_size) for path in image_paths]
    masks = [load_mask(path, target_size) for path in mask_paths]

    images = np.array(images)
    masks = np.array(masks)

    if len(images) == 0 or len(masks) == 0:
        raise ValueError("Loaded images or masks are empty.")

    # Normalize images
    images = images / 255.0
    # Normalize masks (0 or 1 values)
    masks = masks / 255.0

    return images, masks

# Check if images and masks are loaded correctly
image_folder = '/content/drive/MyDrive/crack_project/crack_dataset/train/images'
mask_folder = '/content/drive/MyDrive/crack_project/crack_dataset/train/masks'
images, masks = load_data(image_folder, mask_folder)

print(f"Number of images: {len(images)}")
print(f"Number of masks: {len(masks)}")
print(f"Shape of first image: {images[0].shape if len(images) > 0 else 'No images loaded'}")
print(f"Shape of first mask: {masks[0].shape if len(masks) > 0 else 'No masks loaded'}")
```

```python
import os

def sort_and_trim_files(folder_path, keep_count):
    # 파일 목록 생성
    files = [f for f in os.listdir(folder_path) if f.lower().endswith('.jpg')]

    # 파일 이름 내림차순 정렬
    files.sort(reverse=True)

    # 삭제할 파일 목록
    files_to_delete = files[keep_count:]  # keep_count를 넘는 파일은 삭제

    # 파일 삭제
    for file_name in files_to_delete:
        file_path = os.path.join(folder_path, file_name)

        if os.path.exists(file_path):
            os.remove(file_path)
            print(f"Deleted file: {file_path}")

    print(f"Remaining files: {len(os.listdir(folder_path))}")

# 폴더 경로 설정
image_folder = '/content/drive/MyDrive/crack_project/crack_dataset/train/images'
mask_folder = '/content/drive/MyDrive/crack_project/crack_dataset/train/masks'

# 각 폴더에서 300개만 남기고 나머지 삭제
sort_and_trim_files(image_folder, 300)
sort_and_trim_files(mask_folder, 300)
```

```python
import numpy as np
import cv2
import glob

def load_image(image_path, target_size=(256, 256)):
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f"Image at {image_path} could not be loaded.")
    img = cv2.resize(img, target_size)
    img = img.astype(np.float32) / 255.0
    return img

def load_mask(mask_path, target_size=(256, 256)):
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    if mask is None:
        raise ValueError(f"Mask at {mask_path} could not be loaded.")
    mask = cv2.resize(mask, target_size)
    mask = mask.astype(np.float32) / 255.0
    return mask

def load_data(image_folder, mask_folder, target_size=(256, 256)):
    image_paths = glob.glob(f"{image_folder}/*.jpg")
    mask_paths = glob.glob(f"{mask_folder}/*.jpg")

    if not image_paths:
        raise ValueError(f"No images found in {image_folder}")
    if not mask_paths:
        raise ValueError(f"No masks found in {mask_folder}")

    # Find common file names
    image_files = set(f.lower() for f in os.listdir(image_folder) if f.lower().endswith('.jpg'))
    mask_files = set(f.lower() for f in os.listdir(mask_folder) if f.lower().endswith('.jpg'))
    common_files = image_files.intersection(mask_files)

    if len(common_files) == 0:
        raise ValueError("No common files found between images and masks folders.")

    common_files = list(common_files)

    images = []
    masks = []

    for file_name in common_files:
        img_path = os.path.join(image_folder, file_name)
        mask_path = os.path.join(mask_folder, file_name)

        img = load_image(img_path, target_size)
        mask = load_mask(mask_path, target_size)

        images.append(img)
        masks.append(mask)

    images = np.array(images)
    masks = np.array(masks)

    return images, masks

# Define the folder paths
image_folder = '/content/drive/MyDrive/crack_project/crack_dataset/test/images'
mask_folder = '/content/drive/MyDrive/crack_project/crack_dataset/test/masks'

# Load the data
test_images, test_masks = load_data(image_folder, mask_folder)

# Print the sizes of the data
print(f"Number of test images: {len(test_images)}")
print(f"Number of test masks: {len(test_masks)}")
```

```python
import os
import shutil

# 폴더 경로 설정
source_images_folder = '/content/drive/MyDrive/crack_project/crack_dataset/images'
source_masks_folder = '/content/drive/MyDrive/crack_project/crack_dataset/masks'
destination_folder = '/content/drive/MyDrive/crack_project/crack_dataset/test'

# 생성할 폴더 경로
destination_images_folder = os.path.join(destination_folder, 'images')
destination_masks_folder = os.path.join(destination_folder, 'masks')

# 폴더 생성 (이미 존재할 경우 무시)
os.makedirs(destination_images_folder, exist_ok=True)
os.makedirs(destination_masks_folder, exist_ok=True)

# 파일 목록 생성 (대소문자 구분 없음)
image_files = set(f.lower() for f in os.listdir(source_images_folder) if f.lower().endswith('.jpg'))
mask_files = set(f.lower() for f in os.listdir(source_masks_folder) if f.lower().endswith('.jpg'))

# 공통 파일 찾기
common_files = list(image_files.intersection(mask_files))

# 공통 파일 중 최대 300개 선택
files_to_copy = common_files[:300]

# 파일 복사
for file_name in files_to_copy:
    src_image_path = os.path.join(source_images_folder, file_name)
    src_mask_path = os.path.join(source_masks_folder, file_name)

    dst_image_path = os.path.join(destination_images_folder, file_name)
    dst_mask_path = os.path.join(destination_masks_folder, file_name)

    # 파일 존재 여부 확인 및 복사
    if os.path.exists(src_image_path):
        shutil.copy(src_image_path, dst_image_path)
    else:
        print(f"Image file not found: {src_image_path}")

    if os.path.exists(src_mask_path):
        shutil.copy(src_mask_path, dst_mask_path)
    else:
        print(f"Mask file not found: {src_mask_path}")

    print(f"Copied {file_name} to test/images and test/masks")

print("File copying complete.")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.image import resize

# 모델 로드
model_path = '/content/drive/MyDrive/crack_project/crack_dataset/train/unet_model.h5'
model = load_model(model_path)

# 이미지 전처리 함수
def preprocess_image(image_path, target_size=(256, 256)):
    img = load_img(image_path, target_size=target_size)
    img = img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize to [0, 1] range
    return img

# 예측 수행 함수
def predict_mask(image_path, model):
    img = load_img(image_path)
    original_size = img.size
    preprocessed_img = preprocess_image(image_path)

    pred = model.predict(preprocessed_img)
    pred = (pred > 0.5).astype(np.uint8)  # Thresholding at 0.5

    # 예측 마스크를 원본 이미지 크기로 리사이즈
    original_size_reversed = (original_size[1], original_size[0])  # (height, width) 순서로 변경
    pred_resized = resize(pred[0], original_size_reversed)
    pred_resized = (pred_resized > 0.5).numpy().astype(np.uint8)  # Thresholding after resize
    return pred_resized

# 결과 시각화 함수
def plot_prediction(image_path, prediction):
    img = load_img(image_path)
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title('Input Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(prediction.squeeze(), cmap='gray')
    plt.title('Predicted Mask')
    plt.axis('off')

    plt.show()

# 훼손된 부분의 비중 계산 함수
def calculate_damage_percentage(mask):
    if mask.ndim > 2:  # If mask has multiple channels, sum across the channels
        mask = np.sum(mask, axis=-1)
    total_pixels = mask.size
    damaged_pixels = np.sum(mask)
    damage_percentage = (damaged_pixels / total_pixels) * 100
    return damage_percentage

# 테스트 이미지 경로
test_image_path = '/content/drive/MyDrive/crack_project/crack_dataset/22d959c4.jpg'

# 예측 및 시각화
predicted_mask = predict_mask(test_image_path, model)
plot_prediction(test_image_path, predicted_mask)

# 훼손된 부분의 비중 계산 및 출력
damage_percentage = calculate_damage_percentage(predicted_mask)
print(f"Percentage of damaged area: {damage_percentage:.2f}%")
```
